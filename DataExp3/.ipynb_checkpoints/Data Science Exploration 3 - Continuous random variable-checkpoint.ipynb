## In this data science exploration, we will be performing basic statistical analysis on a real world datasets. The analysis will consist of plotting the empirical distribution of the quantities of interest in the dataset and visualizing how well the empirical distribution matches with continuous probability distributions learned in the class.

```python
## Importing the necessary packages

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

## Indian premier league match dataset: data on matches played in the Indian premier league from 2008-2018. The data has many attributes but in this assignment we will only focus on some of the attributes

```python
## Loading the dataset as a pandas dataframe and printing the header
matches = pd.read_csv('matches.csv')
matches.head()
```

## Plot a un-normalized bar plot of the winning margin in terms of the number of wickets: win by wickets in the horizontal axis and the un-normalized frequency in the vertical axis. Observe that the winning margin can be either in terms of wickets or in terms of runs but not both.

```python

# Start your code here

plt.hist(matches['win_by_wickets'],bins=10)
plt.xlabel("win by wickets")
plt.ylabel("frequency")
plt.show()

# End your code here

```

## Plot a normalized bar plot of the winning margin in terms of the number of wickets: win by wickets in the horizontal axis and the normalized frequency in the vertical axis. Recall, that this will generate a plot of the empirical Probability Density Function (PDF).

```python
# Start your code here

plt.hist(matches['win_by_wickets'],bins=10,density=True)
plt.xlabel("win by wickets")
plt.ylabel("frequency")
plt.show()

# End your code here
```

## Plot the cumulative distribution of the winning margin in terms of the number of wickets: win by wickets in the horizontal axis and cumulative frequency in the vertical axis. Recall, that this will generate a plot of the empirical Cumulative Distribution Function (CDF).

```python
# Start your code here

plt.hist(matches['win_by_wickets'],bins=10,density=True,cumulative=True)
plt.xlabel("win by wickets")
plt.ylabel("cumulative frequency")
plt.show()

# End your code here
```

## Fit a gaussian distribution to the empirical PDF: plot the empirical PDF and the gaussian distribution fitting the emprical PDF on the same plot

```python
# Start your code here

mu=matches['win_by_wickets'].mean()
sigma=matches['win_by_wickets'].std()
x=np.linspace(0,10,100)
y=stats.norm.pdf(x,mu,sigma)
plt.plot(x,y)
plt.hist(matches['win_by_wickets'],bins=10,density=True)
plt.xlabel("win by wickets")
plt.ylabel("frequency")
plt.legend(['normal distribution','histogram'])
plt.show()

# End your code here
```

## Indian premier league deliveries dataset: data on balls bowled in the Indian premier league from 2008-2018. The data has many attributes but in this assignment we will only focus on some of the attributes

```python
## Loading the dataset as a pandas dataframe and printing the header
deliveries = pd.read_csv('deliveries.csv')
deliveries.head()
```

```python
deliveries.columns
```

## Process the dataframe and create a new dataframe with the following structure:

## - Rows correspond to the match ID

## - Column 1 correspond to the total number of runs scored in the match

## - Column 2 correspond to the total number of runs scored in the match by the batsman

## After you have created the dataframe, print the header of the dataframe.

```python
# Start your code here
mod_df=deliveries.groupby('match_id')['batsman_runs','total_runs'].sum()
mod_df.head()
# End your code here
```

## Fitting a gaussian distribution to the total number of runs scored in a match: plot the empirical PDF of the total number of runs scored in a match and the gaussian distribution fitting the emprical PDF on the same plot

```python

# Start your code here
mu=mod_df['batsman_runs'].mean()
sigma=mod_df['batsman_runs'].std()
x=np.linspace(0,max(mod_df['batsman_runs']),100)
y=stats.norm.pdf(x,mu,sigma)
plt.plot(x,y)

plt.hist(mod_df['total_runs'],bins=20,density=True)
plt.xlabel("total runs")
plt.ylabel("frequency")
plt.legend(['fitted normal distribution','histogram'])
plt.show()

# End your code here
```

## Fitting a gaussian distribution to the total number of runs scored by batsman in a match: plot the empirical PDF of the total number of runs scored by batsman in a match and the gaussian distribution fitting the emprical PDF on the same plot

```python
# Start your code here

mu=mod_df['batsman_runs'].mean()
sigma=mod_df['batsman_runs'].std()
x=np.linspace(0,max(mod_df['batsman_runs']),100)
y=stats.norm.pdf(x,mu,sigma)

plt.plot(x,y)
plt.hist(mod_df['batsman_runs'],bins=20,density=True)
plt.xlabel("batsman runs")
plt.ylabel("frequency")
plt.legend(['fitted normal distribution','histogram'])
plt.show()

# End your code here
```
